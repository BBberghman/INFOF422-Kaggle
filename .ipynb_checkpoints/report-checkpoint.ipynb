## 2. Feature selection

First of all it is important to remark that feature selection is relative to a learning algorithm. If you change your learning algorithm you will probably need to change your feature selection. As you need to do this feature selection to get the best model, you can see that the process is in fact cyclic: you do a first feature selection to assess the quality of a model, then given that model you can modify your selected features etc. This is thus an iterative process. Here I will describe the first step using a linear model.

### 2.2 Filter methods

   Filter methods create a subset of features, removing from the whole features set the ones less likely to determine the variable (SalePrice). It is robust to overfitting and effective in computational time. However it might select redundant variables as the interraction between the variables is not taken in consideration. We will consider two filter methods: correlation and mRMR. We can expect that they will give the same set, eventually smaller for mRMR as they process in the same way. The difference is that mRMR also try to avoid redundant variables. 
   
   The correlation method workds as follow:
   
   ```
     Execute the correlation between the dataSet (X) and the variable (Y)
     Order them by descending order, ordering
     Repeat for k = 1:#features
        Select k features of this ordering and compute the CV error
     Select the number of features corresponding to the smallest CV error
        features = ordering[1:argmin(CVerror)]
        ```

````r
source("functions/correlation.R")
source("functions/mRMR.R")
features.corr <- correlation(X, Y)
features.mrmr <- mrmr(X, Y)
````

Now that we've selected criterias by 2 different methods, we will look if they coincide with each other by taking their intersection. mRMR does not only take into account the correlation with the output, but also by avoiding redudant variables. To choose which one we will keep, let's have a look at the lowest CV-error for both methods:

> The CV error is the 10-fold cross validation executed on the model in order to assess the validity and the correctness of a model. It consists on taking 1/10 of the dataset for testing and 9/10 of the dataset for training. The model is then computed given the informaiton of the training set and then assess on the testing set. This is done 10 times, each time changing the portion of data which is testing and training. The CV error is the mean over the 10 tests of the root mean square of the difference of the log. 

### 2.3 Wrapper method

The wrapper method is a cyclic method where a subset of variablse is created and evaluated by the Learning Algorithm, modifying the chosen subset. This is done until the best subset is generated. Recall that this procedure is relative to a Learning Algorithm. Here it is the linear model.

The wrapper method works as follow:


```
Select the feature that gives the smallest CV error amongst the n features
Repeat untill all features selected:
    Select the feature amongst the one left which, along with the one already selected, gives the smallest CV error 
```
One feature is thus added each step. To choose the number of features, one needs to take the smallest CV error.
It's kind of a greedy algorithm, each step adding one feature by the best improvement procedure.

````r
source("functions/wrapperLM.R")
features.wrapper <- wrapperLM(X, Y)
````

### 2.4 Hybrid method
This method is a mix of a filter method and a wrapper one: the filter method is used to select a first "big" set of features, that is then refined by the wrapper method. This gives us the possibility to use advantages of both method to get a good subset in a relatively correct computational time. Indeed the wrapper method quite demanding in term of computation, reducing the subset first could help this problem.

````r
features.wrapper.mrmr <- wrapperLM(X[,features.mrmr],Y)
````

### 3.2 Neural network

To optimize the parameter $size$, as always when you're optimizing something, you first try to find the magnitude order by doing a rapid optimization by large steps and then when you know approximately where your optimum is you refine your research to have the exact value of the parameter that is optimized.

> Remark: the neural network needs normalized data, unlike the linear model and the decision tree. When normalization is done during cross-validation, one thing to be careful about is to not normalized by the full set of data but by the mean and standard deviation of the training set. Indeed both sets must follow the same normalization in order to get matching model.

#### Parameter *size*
An analysis of the number of neurones is done. It is firstly done on the whole data to have the empirical error.

````r
empirical_error <- numeric(21)

# scaling parameters
X.sc <- data.frame(scale(X.feature))
X.sd <- attr(X.sc, 'scaled:scale')
X.mean <- attr(X.sc, 'scaled:center')

Y.sc <- scale(Y)
Y.sd <- attr(Y.sc, 'scaled:scale')
Y.mean <- attr(Y.sc, 'scaled:center')

DS <- cbind(X.sc, SalePrice=Y.sc)

for (i in 1:21){
    model.nn <- nnet(SalePrice~.,DS, size = i, linout=T, maxit = 500, trace = F) 
    
    Y.hat.sc <- predict(model.nn,X.sc)
    Y.hat <- Y.hat.sc*Y.sd + Y.mean
    Y.unsc <- Y.sc*Y.sd + Y.mean
    
    # prevent log error
    Y.hat[Y.hat <= 0] <- 1    
    Y.unsc[Y.unsc <= 0] <- 1
    
    empirical_error[i] <- sqrt( mean( (log(Y.hat)-log(Y.unsc))^2 ))
    message(paste("Neural network: size = ", i, "; Empirical error =",round(empirical_error[i],digits=4)))
}
````

````r
plot(c(1:21),empirical_error[1:21], xlab = "Number of neurones", ylab = "Empirical error")
````

#### Cross validation

## 4. Ensemble techniques 

Ensemble techniques consist on combinating models together in order to get an estimate of the *SalePrice*. The combination is done by averaging the *SalePrice* estimated by *R* models. In a first part we will only combinate models that are of the same type. After that we will try to combine 2 different types of models: linear and neural network. The ensemble techniques are usefull to decrease the variance of your model as seen in the Lecture.  The combinations are tested on the 2 sets of features: the one obtained by the linear model Learning algorithm, and the one obtained by the neural network Learning algorithm (just obtained before).

The pseudo-code is the following: 

```
R <- number of models to average
Y.ts <- testing set
Y.tr <- training set

for r in c(1:R)
    resample the training set
    create a model with the resampled training set
    Y[r] <- prediction of Y.ts with the model

average the results obtained by the prediction of the R models
```

````r
#read and prep testing set
data <- data.type("input/test.csv")
data$numeric <- data$numeric[,2:37]
variable_to_keep <- c("Neighborhood","CentralAir")
data_factor_onehot <- dummy.data.frame(data$factor[,variable_to_keep], sep="_")
data.extended <- cbind(data$numeric,data_factor_onehot)
data.nona <- data.frame(apply(data.extended,2,replace_na_with_mean_value)) 
correlation.matrix <- cor(data.nona)
correlation.matrix[upper.tri(correlation.matrix)] <- 0
diag(correlation.matrix) <- 0
data.uncorrelated <- data.nona[,!apply(correlation.matrix,2,function(x) any(abs(x) > 0.99))]
X <- data.uncorrelated
 
# first set
features.wrapper <- cbind(16,6,12,52,21,7,9,1,5,50,3,58,57,43,24,27,42,40,32,60,45,61,8,17,19,22,25,30,28,15,36,56,35,59,41,20,38,49,47,31,27,23,55,11,10,53,51,33,29,39,54,62,14,12,18,46,2,44,48,34,26,4)
features.wrapper <- features.wrapper[1:33]
features.selected <- features.wrapper
X.features.testing <- X[,features.selected]

# second set
features.mrmr.nn.saved <- c(4, 49, 16, 26, 44, 12, 7, 54, 27, 19, 46, 13, 6, 23, 24, 40, 8, 25, 52, 9, 14, 55, 50, 28, 2, 20, 58, 29, 30, 3, 47, 62, 22, 17, 11, 39, 5, 57, 32, 60, 59, 21, 43, 1, 42, 48, 61, 33, 41, 51, 35)
features.wrapper.nn <- c(1,3,13,20,30,37,35,5,48,6,44,49,7,27,9,17,42,26,47,15,38,28,8,40,51,22)
temp <- X[,features.mrmr.nn.saved]
X.featuresNN.testing <- temp[,features.wrapper.nn[1:22]]
````

````r
# predictions
X.testing <- X.features.testing
Y.training <- Y
X.training <- X.features.training
N<-nrow(X.testing)   
n<-ncol(X.training)
size.CV<-floor(N/10)
R<-10
CV.err<-numeric(10)    
    
i.tr<-1:N 
SalePrice.R <- matrix(0,nrow=N,ncol=4*R)

for (r in 1:floor(R/4)){
    # training
    i.tr.resample<-sample(i.tr,rep=T)
    X.tr<-X.training[i.tr.resample,]
    Y.tr<-Y.training[i.tr.resample]

    DS<-cbind(X.tr,SalePrice=Y.tr)    
    model <- lm(SalePrice~.,DS)

    SalePrice.hat<- predict(model,X.testing)  
    SalePrice.hat[SalePrice.hat <= 0] <- 1 
    SalePrice.R[,r] <- SalePrice.hat
}  
for (r in floor(R/4)+1:R) {
    # training
    i.tr.resample<-sample(i.tr,rep=T)
    X.tr<-X.training[i.tr.resample,]
    Y.tr<-Y.training[i.tr.resample]     

    # scaling parameters
    X.sc <- scale(X.tr)
    X.sd <- attr(X.sc, 'scaled:scale')
    X.mean <- attr(X.sc, 'scaled:center')
    Y.sc <- scale(Y.tr)
    Y.sd <- attr(Y.sc, 'scaled:scale')
    Y.mean <- attr(Y.sc, 'scaled:center')
    X.sc <- data.frame(X.sc)

    X.ts.sc <- t(apply(sweep(X.testing,2,X.mean,"-"), 1, function(x) x/X.sd)) 
    X.ts.sc <- data.frame(X.ts.sc)

    # training = dataset
    DS <- cbind(X.sc, SalePrice=Y.sc)
    DS<- DS[,colSums(!is.na(DS)) > 0]  # if one column has only NAs due to std = 0 

    model.nn <- nnet(SalePrice~.,DS, size = 2, linout=T, maxit = 500, trace=F)
    Y.hat.sc <- predict(model.nn,X.ts.sc)
    SalePrice.hat <- Y.hat.sc*Y.sd + Y.mean

    SalePrice.hat[SalePrice.hat <= 0] <- 1    
    SalePrice.R[,r] <- SalePrice.hat
}

SalePrice<-apply(SalePrice.R,1,mean)
    
Id <- read.csv("input/test.csv")$Id
write.table(data.frame(Id,SalePrice), file="output/predictions.csv", sep=",", row.names=F)    
````
